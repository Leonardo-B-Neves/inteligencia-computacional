```python
import numpy as np
import matplotlib.pyplot as plt

# Dados
dados = {
    'T': ['T1', 'T2', 'T3', 'T4', 'T5', 'T6', 'T7', 'T8'],
    'Temperatura': [10, 8, 6, 5, 20, 22, 23, 18],
    'Umidade': [20, 13, 15, 11, 26, 26, 27, 23],
    'Flor': [1, 0, 1, 1, 0, 0, 1, 0]
}

# Parâmetros do perceptron
taxa_aprendizado = 0.1
num_epocas = 100

# Inicialização dos pesos e bias
pesos = np.random.rand(2)  # Inicializando aleatoriamente os pesos
bias = np.random.rand()
```

```python
# Função de ativação (degrau)
def degrau(soma):
    return 1 if soma >= 0 else 0
```

```python
# Treinamento do perceptron
for epoca in range(num_epocas):
    erro_total = 0
    for i in range(len(dados['T'])):
        # Entradas
        entrada = np.array([dados['Temperatura'][i], dados['Umidade'][i]])

        # Saída desejada
        saida_desejada = dados['Flor'][i]

        # Soma ponderada das entradas
        soma = np.dot(entrada, pesos) + bias

        # Saída do perceptron
        saida = degrau(soma)


        # Cálculo do erro
        if saida_desejada != saida:
          print("rodada",epoca)
          print(dados['T'][i])
          print(saida)
          print(saida_desejada)
          print("")
          erro = 1
        else:
          erro = 0

        # Atualização dos pesos e bias
        pesos += taxa_aprendizado * erro * entrada
        bias += taxa_aprendizado * erro

        erro_total += erro

    # Verifica se todos os exemplos foram classificados corretamente
    if erro_total == 0:
        print(f"Perceptron treinado com sucesso após {epoca+1} épocas.")
        break
    elif epoca == num_epocas - 1:
        print("O perceptron não convergiu após o número máximo de épocas.")
```
